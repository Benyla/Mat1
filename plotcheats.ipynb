```python
#Plotter en 3d surface og enhedsnormalvektorfelt i et og samme plot
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the parameters
u = np.linspace(0, np.pi, 50)
v = np.linspace(-np.pi, np.pi, 50)

# Create a meshgrid from the parameters
u, v = np.meshgrid(u, v)

# Define the surface
x = 3 * np.sin(u) * np.cos(v)
y = 2 * np.sin(u) * np.sin(v)
z = 2 * np.cos(u)

# Define the surface normal vector field
Nx = 4 * np.sin(u) ** 2 * np.cos(v)
Ny = 6 * np.sin(u) ** 2 * np.sin(v)
Nz = 6 * np.sin(u) * np.sin(v) ** 2 * np.cos(u) + 6 * np.sin(u) * np.cos(u) * np.cos(v) ** 2

# Normalize the surface normal vector field to length 1
N_mag = np.sqrt(Nx ** 2 + Ny ** 2 + Nz ** 2)
mask = N_mag != 0  # Create a mask for non-zero magnitudes
Nx_norm = np.zeros_like(Nx)
Ny_norm = np.zeros_like(Ny)
Nz_norm = np.zeros_like(Nz)
Nx_norm[mask] = Nx[mask] / N_mag[mask]
Ny_norm[mask] = Ny[mask] / N_mag[mask]
Nz_norm[mask] = Nz[mask] / N_mag[mask]

# Create a figure and plot the surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='jet')

# Plot the surface normal vector field with reduced vectors and length 1
stride = 4
ax.quiver(x[::stride, ::stride], y[::stride, ::stride], z[::stride, ::stride],
          Nx_norm[::stride, ::stride], Ny_norm[::stride, ::stride], Nz_norm[::stride, ::stride], length=1)

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Flade med enhedsnormalvektorfelt')

# Show the plot
plt.show()
```
